#Requisitos 02

##Objetivo:
- Para cada uma das redes escolhidas, fazer um gráfico bipartido sobre a assortatividade em relação ao grau dos nós da rede.

```python
# Importando bibliotecas.
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

# Networks with ground-truth communities (com-Youtube)

```python
# Fazendo os download da rede.
!wget https://snap.stanford.edu/data/bigdata/communities/com-youtube.ungraph.txt.gz

# Cria um grafo direcionado (DiGraph) a partir dos dados do arquivo baixado
com_youtube = nx.read_edgelist('com-youtube.ungraph.txt.gz', create_using=nx.DiGraph(), data=False)

# average degree of neighbors
degree, avg_neigh_degree = zip(*nx.average_degree_connectivity(com_youtube).items())

# convert to list
degree = list(degree)
avg_neigh_degree = list(avg_neigh_degree)

# Define o estilo do gráfico e ria uma figura e um eixo para o gráfico com dimensões
plt.style.use("fivethirtyeight")
fig, ax = plt.subplots(1,1,figsize=(12,8))

# Gera um gráfico de dispersão com grau no eixo x e grau médio dos vizinhos no eixo y
sns.regplot(x=degree,y=avg_neigh_degree,ax=ax)

# Define o rótulo do eixo x e do y
ax.set_xlabel("Node Degree")
ax.set_ylabel("Average neigbhor degree")

# Define os limites do eixo x
ax.set_xlim(0, 5000)

# Define o título do gráfico
ax.set_title("Degree Assortativity com_youtube")

# Exibe o gráfico gerado
plt.show()
```
